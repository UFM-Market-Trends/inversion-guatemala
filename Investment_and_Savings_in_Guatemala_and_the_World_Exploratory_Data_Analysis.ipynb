```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use("fivethirtyeight")
```

```python
# Download copy of Penn World Table version 10.01
pwt10001_url = "https://github.com/UFM-Market-Trends/inversion-guatemala/raw/main/data/pwt1001.xlsx"

pwt10001_df = pd.read_excel(pwt10001_url, sheet_name=2)
```

```python
for code in country_codes:
  years = pwt10001_df[pwt10001_df.countrycode == code]["year"]
  # Output-side real GDP at chained PPPs (in mil. 2017US$)
  rgdpo = pwt10001_df[pwt10001_df.countrycode == code]["rgdpo"]
  pop = pwt10001_df[pwt10001_df.countrycode == code]["pop"]
  gdppc = rgdpo / pop

  plt.plot(years, gdppc)

plt.show()
```

```python
# Initialize a dictionary to store average GDP per capita for each region
regions_avg_gdppc = {}

for region, region_codes in regions.items():
    # Filter the DataFrame for the countries in the current region
    region_df = pwt10001_df[pwt10001_df['countrycode'].isin(region_codes)]

    # Group by year and calculate the mean GDP per capita
    # avg_gdppc_by_year = region_df.groupby('year').apply(lambda df: (df['rgdpo'] / df['pop']).mean())
    # Actually, don't use simple mean, use average weighted by population size

    # Calculate the total GDP and total population for each year in the region
    total_gdp_and_pop_by_year = region_df.groupby('year').apply(lambda df: pd.Series({
        'total_rgdpo': df['rgdpo'].sum(),
        'total_pop': df['pop'].sum()
    }))

    # Calculate the weighted average GDP per capita for the region
    avg_gdppc_by_year = total_gdp_and_pop_by_year['total_rgdpo'] / total_gdp_and_pop_by_year['total_pop']

    # Store the result in the dictionary
    regions_avg_gdppc[region] = avg_gdppc_by_year

plt.figure(figsize=(7, 7))  # Width, height in inches

# Plot the average GDP per capita for each region
for region, avg_gdppc_by_year in regions_avg_gdppc.items():
    plt.plot(avg_gdppc_by_year.index, avg_gdppc_by_year, label=region)

plt.legend()
plt.title("GDP Per Capita by Region (1950-2019)")
plt.ylabel("GDP Per Capita (in 2017 USD)")
caption_text = ("This graph displays the regional GDP per capita from 1950 to 2019,"
                  "\ncalculated by dividing the total output-side real GDP at chained PPPs"
                  "\nby the total population for each region. "
                  "The data is sourced from the \nPenn World Table 10.01. Chart by @JoseAlvarezC97.")

# Add the caption
plt.figtext(0.5, -0.08, caption_text, wrap=True, horizontalalignment='center', fontsize=10)

plt.show()
```

```python
# Let's define a reusable function

def get_regions_avg_variable_percapita(var="rgdpo", pop="pop"):
  # Initialize a dictionary to store average <variable> per capita for each region
  regions_avg_variable_percapita = {}

  for region, region_codes in regions.items():
      # Filter the DataFrame for the countries in the current region
      region_df = pwt10001_df[pwt10001_df['countrycode'].isin(region_codes)]

      # Calculate the total <variable> and total population for each year in the region
      total_variable_and_pop_by_year = region_df.groupby('year').apply(lambda df: pd.Series({
          f'total_{var}': df[var].sum(),
          'total_pop': df[pop].sum()
      }))

      # Calculate the weighted average GDP per capita for the region
      avg_variable_by_year = (total_variable_and_pop_by_year[f'total_{var}']
                              / total_variable_and_pop_by_year['total_pop'])

      # Store the result in the dictionary
      regions_avg_variable_percapita[region] = avg_variable_by_year

  return regions_avg_variable_percapita


def plot_regions_avg_variable_percapita(regions_avg_variable_percapita,
                                        var_name="GDP",
                                        plot_caption=True,
                                        figsize=(7, 7)):
  plt.figure(figsize=figsize)  # Width, height in inches

  # Plot the average GDP per capita for each region
  for region, avg_variable_by_year in regions_avg_variable_percapita.items():
      plt.plot(avg_variable_by_year.index, avg_variable_by_year, label=region)

  plt.legend()
  plt.title(f"{var_name} Per Capita by Region (1950-2019)")
  plt.ylabel(f"{var_name} Per Capita (in 2017 USD)")
  if plot_caption == True:
    caption_text = (f"This graph displays the regional {var_name} per capita from 1950 to 2019,"
                      f"\ncalculated by dividing the total {var_name}"
                      "\nby the total population for each region. "
                      "The data is sourced from the \nPenn World Table 10.01. Chart by @JoseAlvarezC97.")

    # Add the caption
    plt.figtext(0.5, -0.08, caption_text, wrap=True, horizontalalignment='center', fontsize=10)
    plt.show()
```

```python
regions_avg_gdppc_test = get_regions_avg_variable_percapita()
plot_regions_avg_variable_percapita(regions_avg_gdppc_test)
```

```python
regions_avg_cnpc_test = (
    get_regions_avg_variable_percapita(var="cn",
                                       # pop is technically not per "capita",
                                       # but per "worker", so caption and
                                       # title would have to be adjusted in
                                       # formal presentation
                                       pop="emp"
                                       )
    )
plot_regions_avg_variable_percapita(regions_avg_cnpc_test,
                                    var_name="Capital Stock (current PPPs)",
                                    plot_caption=False,
                                    figsize=(8,8))
```
